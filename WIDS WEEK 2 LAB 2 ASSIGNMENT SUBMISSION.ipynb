```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def g(x1,x2):
  return (x1+49)**2 + (x2-36)**2

def f(x1,x2):
  return 256*((x2-x1**2)**2) + (2-x1)**2

def h(x1,x2):
  return x1**2 + x2**2 + 9
```

```python
A = np.array([[1, 0], [0, 1]])
b = np.array([98, -72])
c = 49**2 + 36**2
def g_matrix(x1,x2):
    x = np.array([x1,x2])
    return x.T @ A @ x + b.T @ x + c
```

```python
def grad_g(x1,x2):
  dgdx1=2*(x1+49)
  dgdx2=2*(x2-36)
  return [dgdx1,dgdx2,((dgdx1**2)+(dgdx2**2))**0.5]

def grad_f(x1,x2):
  dfdx1=-1024*x1*(x2-x1**2)-2*(2-x1)
  dfdx2=512*(x2-x1**2)
  return [dfdx1,dfdx2,((dfdx1**2)+(dfdx2**2))**0.5]

def grad_h(x1,x2):
  dhdx1=2*x1
  dhdx2=2*x2
  return [dhdx1,dhdx2,((dhdx1**2)+(dhdx2**2))**0.5]

```

```python
def T(p):
  return 10**(-p)




```

```python
tol=[]
it=[]
x1,x2=36,-49
print("TolVals Iterations")
x1_h=[x1]
x2_h=[x2]
for p in range(1,16):
  count=0
  while (grad_g(x1,x2)[2])>T(p):
    i=sp.symbols('i')
    v=g(x1-(i*(grad_g(x1,x2)[0])),x2-(i*(grad_g(x1,x2)[1])))
    dv_dz=sp.diff(v,i)
    c_p = sp.solve(dv_dz, i)
    x1=x1-(c_p[0]*(grad_g(x1,x2)[0]))
    x2=x2-(c_p[0]*(grad_g(x1,x2)[1]))
    count+=1
    x1_h.append(x1)
    x2_h.append(x2)
  tol.append(10**(-p))
  it.append(count)
  print(10**(-p),"      ",count)

print("Min value of g(x1,x2) is ",g(x1,x2))
print("Minimizer x1 is ",x1,"\nMinimizer x2 is ",x2)

plt.plot(tol,it, marker='o', linestyle='-', color='b', label='Line Plot')
plt.xlabel('Tolerance Values')
plt.ylabel('Number of Iterations')
plt.title('Relationship between Tolerance and No. of Iterations')
plt.grid(True)
plt.show()

x1 = np.linspace(-100, 100, 500)
x2 = np.linspace(-100, 100, 500)
x1_grid, x2_grid = np.meshgrid(x1, x2)

plt.figure(figsize=(8, 8))
contour = plt.contour(x1_grid, x2_grid, g(x1_grid,x2_grid), levels=np.logspace(-1, 3, 10), cmap="viridis")
plt.colorbar(contour, label="Function value")
plt.title("Level Sets of the Function")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(alpha=0.3)
plt.scatter(x1_h, x2_h, marker='o', color='red', label='Optimization Path')
plt.show()
```

```python
def k(x):
    n = len(x)
    return sum((x[i] - (i + 1)**2)**2 for i in range(n))

def grad_k(x):
    n = len(x)
    return np.array([2 * (x[i] - (i + 1)**2) for i in range(n)])

def gradient_descent(n, learning_rate=0.1, tolerance=1e-6, max_iterations=10000):
    x = np.random.randn(n)
    origin_x = x
    history = [k(x)]
    iterations = 0

    while iterations < max_iterations:
        grad = grad_k(x)
        grad_norm = np.linalg.norm(grad)
        if grad_norm <= tolerance:
            break
        x = x - learning_rate * grad
        history.append(k(x))
        iterations = iterations + 1

    return origin_x, x, history, iterations

n = 10
learning_rate = 0.1
tolerance = 1e-6
max_iterations = 10000


initial_x, x_min, history, iterations = gradient_descent(n, learning_rate, tolerance, max_iterations)

print("Inital x values", initial_x)
print("Optimized x:", x_min)
print("Minimum value of g(x):", k(x_min))
print("Number of iterations:", iterations)


```

```python
x1, x2 = sp.symbols('x1 x2')

dz_dx1 = sp.diff(h(x1, x2), x1)
dz_dx2 = sp.diff(h(x1, x2), x2)
d2z_dx12 = sp.diff(dz_dx1, x1)
d2z_dx22 = sp.diff(dz_dx2, x2)
d2z_dx1x2 = sp.diff(dz_dx1, x2)

dz_dx1_func = sp.lambdify((x1, x2), dz_dx1, 'numpy')
dz_dx2_func = sp.lambdify((x1, x2), dz_dx2, 'numpy')
d2z_dx12_func = sp.lambdify((x1, x2), d2z_dx12, 'numpy')
d2z_dx22_func = sp.lambdify((x1, x2), d2z_dx22, 'numpy')
d2z_dx1x2_func = sp.lambdify((x1, x2), d2z_dx1x2, 'numpy')

def grad(x1, x2):
    return np.array([dz_dx1_func(x1, x2), dz_dx2_func(x1, x2)])

def hessian(x1, x2):
    return np.array([[d2z_dx12_func(x1, x2), d2z_dx1x2_func(x1, x2)],
                     [d2z_dx1x2_func(x1, x2), d2z_dx22_func(x1, x2)]])

def norm(x1, x2):
    return np.linalg.norm(grad(x1, x2))

def tolerance(p):
    return 10**(-p)

def algo(cp_x1, cp_x2):
    tablelist = []
    x1_h=[cp_x1]
    x2_h=[cp_x2]
    Tvals = []
    Iterations = []

    for i in range(1, 21):
        Tlevel = tolerance(i)
        count = 0
        while norm(cp_x1, cp_x2) > Tlevel:
            g = grad(cp_x1, cp_x2)
            hess = hessian(cp_x1, cp_x2)
            if np.linalg.det(hess) == 0:
                print("Hessian is singular. Exiting iteration.")
                break

            h_inv = np.linalg.inv(hess)
            step = np.dot(h_inv, g)
            cp_x1 -= step[0]
            cp_x2 -= step[1]
            count += 1
            x1_h.append(cp_x1)
            x2_h.append(cp_x2)


        tablelist.append((Tlevel, count, h(cp_x1, cp_x2), cp_x1, cp_x2))
        Tvals.append(Tlevel)
        Iterations.append(count)

    table_array = np.array(tablelist)

    return cp_x1, cp_x2, h(cp_x1, cp_x2),x1_h, x2_h, Tvals, Iterations, table_array

cp_x1, cp_x2 = 1000,-1000
min_x1, min_x2, min_z,histx1,histx2, Tvalues, Iterations1, t = algo(cp_x1, cp_x2)
print(f"Minimum value of z: {min_z}")
print(f"Minimizing values: x1 = {min_x1}, x2 = {min_x2}")
print("Tolerance | Iterations | Function Value | Minimiser X1 | Minimiser X2 ")
print(t)
print(Tvalues)
print(Iterations1)

x1 = np.linspace(-100, 100, 500)
x2 = np.linspace(-100, 100, 500)
x1_grid, x2_grid = np.meshgrid(x1, x2)

plt.figure(figsize=(8, 8))
contour = plt.contour(x1_grid, x2_grid, h(x1_grid,x2_grid), levels=np.logspace(-1, 3, 10), cmap="viridis")
plt.colorbar(contour, label="Function value")
plt.title("Level Sets of the Function")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(alpha=0.3)
plt.scatter(histx1, histx2, marker='o', color='red', label='Optimization Path')
plt.show()

```
